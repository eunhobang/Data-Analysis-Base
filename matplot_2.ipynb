```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10,0.1)
y1 = 0.3*(x-5)**2+1
y2 = -1.5*x+3
y3 = np.sin(x)**2
y4 = 10*np.exp(-x)+1
```

```python
plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)

plt.show()
```

```python
x = np.linspace(-4, 4, 100)
y1 = x**3
y2 = 10*x**2-2

plt.plot(x,y1,x,y2)
plt.show()
print("y1과 y2가 만나는 부분이 잘 보이지 않음")
```

```python
x = np.linspace(-4, 4, 100)
y1 = x**3
y2 = 10*x**2-2

plt.plot(x,y1,x,y2)
plt.xlim(-1,1)
plt.ylim(-3,3)
plt.show()
print("출력범위를 지정해서 관심 영역 확대")
```

### 그래프 꾸미기

```python
x = np.arange(0,5,1)
y1 = x
y2 = x+1
y3 = x+2
y4 = x+3
```

```python
print("노옵션")
plt.plot(x,y1,x,y2,x,y3,x,y4)
plt.show()
```

```python
print("선 색깔 옵션 지정")
plt.plot(x,y1,'m',x,y2,'y',x,y3,'k',x,y4,'c')
plt.show()
```

```python
print("선 스타일 옵션 지정")
plt.plot(x,y1,'-',x,y2,'--',x,y3,':',x,y4,'-.')
plt.show()
```

```python
print("마커 옵션 지정")
plt.plot(x,y1,'o',x,y2,'^',x,y3,'s',x,y4,'d')
plt.show()
```

```python
print("선, 마커 혼합")
plt.plot(x,y1,'>--r',x,y2,'s-g',x,y3,'d:b',x,y4,'-.Xc')
plt.show()
```

## 라벨, 제목, 격자, 범례, 문자열 표시

```python
x = np.arange(-4.5,5,0.5)
y = 2*x**3
```

```python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph title')
plt.grid(True) # 그래프 격자 추가
plt.show()
```

```python
print('범례추가')
x = np.arange(0,5,1)
y1 = x
y2 = x+1
y3 = x+2
y4 = x+3

plt.plot(x,y1,'>--r',x,y2,'s-g',x,y3,'d:b',x,y4,'-.Xc')
plt.legend(['data1','data2','data3','data4'])
plt.show()
```

```python
print('범례 위치 문자열 옵션')
x = np.arange(0,5,1)
y1 = x
y2 = x+1
y3 = x+2
y4 = x+3

plt.plot(x,y1,'>--r',x,y2,'s-g',x,y3,'d:b',x,y4,'-.Xc')
plt.legend(['data1','data2','data3','data4'], loc='lower right')
plt.show()
```

```python
print('범례 위치 숫자 옵션')
x = np.arange(0,5,1)
y1 = x
y2 = x+1
y3 = x+2
y4 = x+3

plt.plot(x,y1,'>--r',x,y2,'s-g',x,y3,'d:b',x,y4,'-.Xc')
plt.legend(['data1','data2','data3','data4'], loc=4)
plt.show()
```

## 한글표시
그래프에서 한글 표시하려면 한글 폰트로 지정해야 됨(디폴트:sans-serif)

matplotlib.rcParams['font.family'] = '폰트이름'


matplotlib.rcParams['axes.unicode_minus'] = False
- 한글 폰트를 지정한 후에 그래프에서 마이너스(-) 폰트가 깨지는 것을 해결

```python
import matplotlib.font_manager
```

```python
# 가지고 있는 폰트 확인하기
font_list = matplotlib.font_manager.get_fontconfig_fonts()
font_names = [matplotlib.font_manager.FontProperties(fname=fname).get_name() for fname in font_list]

f = open('C:\Windows\Fonts','w')
for font_name in font_names:
    f.write(font_name+'\n')
f.close
```

```python
import matplotlib
```

```python
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
print('범례 위치 숫자 옵션')
x = np.arange(0,5,1)
y1 = x
y2 = x+1
y3 = x+2
y4 = x+3

plt.plot(x,y1,'>--r',x,y2,'s-g',x,y3,'d:b',x,y4,'-.Xc')
plt.legend(['data1','data2','data3','data4'], loc='best')
plt.xlabel('X축')
plt.ylabel('Y축')
plt.grid()

plt.show()
```
