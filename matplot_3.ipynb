```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

```python
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
height = [165, 177, 180, 187, 160, 155, 146]
weight = [62, 86, 79, 104, 45, 50, 35]
```

```python
plt.scatter(height, weight)
plt.xlabel('height(m)')
plt.ylabel('weight(Kg)')
plt.title('Height&Weight')
plt.grid()
```

```python
print('마커 크기와 색 추가')
plt.scatter(height, weight, s=500, c='r') 
plt.xlabel('height(m)')
plt.ylabel('weight(Kg)')
plt.title('Height&Weight')
plt.grid()
```

```python
print('데이터별 마커 크기와 색깔 지정')

size = 100*np.arange(1,8)
colors = ['r','g','b','c','y','k','m']

plt.scatter(height, weight, s=size, c=colors) 
# plt.xlabel('height(m)')
# plt.ylabel('weight(Kg)')
# plt.title('Height&Weight')

print('지역별 인구 밀도, 질병 발생률 등 직관적으로 그래프를 나타낼 수 있임')
```

```python
city = ['서울', '광주', '대구','대전','인천','부산','울산']

# 위도 & 경도
lat = [36.56, 37.18, 35.19, 34.14, 35.15, 37.11, 34.18]
lon = [124.79, 126.83, 127.38, 128.60, 129.31, 126.77, 126.78]

# 인구 밀도
pop_den = [16154, 2751, 2895, 2790, 1099, 4454, 2995]

size = np.array(pop_den)*0.2
colors = ['r','g','b','c','y','k','m']

plt.scatter(lon,lat,s=size, c=colors, alpha=0.5)
plt.xlabel('경도')
plt.ylabel('위도')
plt.title('지역별 인구 밀도(2017)')

for x,y, name in zip(lon, lat, city):
    plt.text(x,y,name)

plt.show()
```
